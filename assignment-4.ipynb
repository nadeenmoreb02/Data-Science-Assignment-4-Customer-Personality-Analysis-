### 7. Select two categorical features and verify if the data is balanced between the categories of each feature

#### First Numerical Feature : Education
the result below show that they are not balanced since their frequencies are not the same or even close 

```python
# education feature 
first_categorical_feature = 'Education'


# frequency of categories for each feature
category_counts1 = df[first_categorical_feature].value_counts()
print("Frequency of categories for", first_categorical_feature, ":")
print(category_counts1)

# balance check (if result is one then it is balanced otherwise no)
balance_feature1 = category_counts1.min() / category_counts1.max()

print("Balance for", first_categorical_feature, ":", balance_feature1)

# acceptable balance range 
if(balance_feature1 > 1.25 or balance_feature1 < 0.78):
    print("not balanced")
else:
    print("balanced")
```

### 8. Based on your intuition, select two features that you think they could be correlated. Verify your assumption using both visualization and the correlation coefficient approaches

Based on my intuition, High incomes can be associated with increase in food, particulary meat consumption, hence income level might influence spending behavior on meat products. 

the features 'Income' and 'MntMeatProducts' in the data set maybe correlated, therefore, I analyzed their correlation using the methods : 

1. visualization 

2. correlation coeffecient

### Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt


# select numerical features
numerical_features = df.select_dtypes(include='number')

# correlation matrix
corrmat = numerical_features.corr()

# plot
plt.figure(figsize=(20, 20))
sns.heatmap(corrmat, annot=True, cmap='coolwarm', center=0)

# title and layout
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()

#display plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature1 = 'Income'
feature2 = 'MntMeatProducts'

# 1. Visualization: Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=feature1, y=feature2)
plt.title('Scatter Plot of {} vs {}'.format(feature1, feature2))
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.grid(True)
plt.show()

# 2. Correlation coefficient
correlation_coefficient = df[[feature1, feature2]].corr().iloc[0, 1]

print("Correlation coefficient between {} and {}: {:.2f}".format(feature1, feature2, correlation_coefficient))
```

the **scatter plot** indicates a positive relation/correlation between the varibles, yet not very strog

The **correlation coefficient** between Income and Amount spent on Meat Products (MntMeatProducts) is 0.58, this is a moderately positive linear relationship between these 2 variables. hence, it is a moderate strength of the relationship, meaning that it can still indicate that customers with higher incomes spend more on meat products. In general, the correlation coefficient supports the intuition these variables are correlated. 

### 9. Are there features that can be used as index?

yes, the ID, because it has unique values, although uniqueness is not a requirement here but it makes it a good candidate since it is a large data set and makes it easier to retrieve, refer or manipulate data.

```python
# uniqueness of 'ID' checking by compare the number of unique id rows to the total number of rows 
unique_id_count = df['ID'].nunique()
total_records = len(df)

#set ID as index if it is unique
if unique_id_count == total_records:
    print("'ID' can be used as an index since it has unique values for each record.")
    # Set 'ID' column as the index
    df.set_index('ID', inplace=True)

else:
    print("'ID' cannot be used as an index because it does not have unique values for each record.")
```
